```python
class SinglyLinkedList:
    """LIFO Stack implementation using a singly linked list for storage."""
    #-------------------------- nested Node class --------------------------
    
    class Node:
        """Lightweight, nonpublic class for storing a singly linked node."""
        __slots__ = '_element', '_next'
        
        def __init__(self, element, next):        
            self._element = element               
            self._next = next

        def setNext(self, next):
            self._next = next
    
        def getNext(self):
            return self._next
  
        def setElement(self, element):
            self.element = _element
    
        def getElement(self):
            return self._element

    def __init__(self, head = None, size = 0):
        """Create an empty list."""
        self._head = head                      
        self._size = size
    
    def getSize(self):
        """Return the number of elements in the list."""
        return self._size
    
    def isEmpty(self):
        """Return True if the list is empty."""
        return self._size == 0
    
    def add (self, element):
        """Add element element to the end of the list."""
        node = self.Node(element, None)
        if self.isEmpty():       
            self._head = node
        else:
            cursor = self._head                 
            while cursor.getNext() is not None:
                cursor = cursor.getNext()
            cursor.setNext(node)
        print('Adding # ', self._size, ' element: ', node.getElement())
        self._size += 1
        
    def remove(self):
        """Remove the element at the head of the list. Raise Empty if the list is empty."""
        if self.isEmpty():
          raise Empty('List is empty')
        self._size -= 1
        out = self._head.getElement()
        self._head = self._head.getNext()
        print('Removed ', out, ' from head')
        return out

    def showList(self):
        """Display the contents of the list, from head to tail. Raise Empty exception if the List is empty. """
        cursor = self._head
        while cursor:
            print(cursor.getElement())
            cursor = cursor.getNext()
        return
    
    def search(self, key):
        """ Search the list using 'key. """        
        cursor = self._head
        while cursor:
            if key == cursor.getElement():
                return True
            cursor = cursor.getNext()
        return False
    
    # --------- Methods added by Hasan Zahid for assignment -----------
    def insertAfter(self, index, value):
        node = self.Node(value, None)
        if self.isEmpty():
            self._head = node
        else:
            if index == 1:
                print(f'Adding at index {index}: {node.getElement()}')
                node.setNext(self._head)
                self._head = node
            else:    
                i = 1
                cursor = self._head
                while i < index-1 and cursor.getNext() is not None:
                    cursor = cursor.getNext()
                    i+=1
                    
#                 print(f'Adding at index {index}: {node.getElement()}')  this also works but below if-else is more readable
#                 node.setNext(cursor.getNext())
#                 cursor.setNext(node)

                if cursor.getNext() is None:
                    print(f'index={index} is out of bounds, Adding to Tail')
                    cursor.setNext(node)
                else:
                    print(f'Adding at index {index}: {node.getElement()}')
                    node.setNext(cursor.getNext())
                    cursor.setNext(node)
        self._size+=1
        
        
    def insertElt(self, value, indicator, element):
        node = self.Node(value, None)
        if self.isEmpty():
            self._head = node
        else:
            if indicator == 'A':
                cursor = self._head
                while cursor.getNext() is not None:
                    if cursor.getElement() == element:
                        break
                    cursor = cursor.getNext()
                if cursor.getNext() is None:
                    print('new node element not found in this SLL, adding to Tail')
                    cursor.setNext(node)
                else:
                    print(f'Inserting After {element}')
                    node.setNext(cursor.getNext())
                    cursor.setNext(node)
            else:
                cursor = self._head
                if cursor.getElement() == element:
                    print('Element found at head')
                    node.setNext(self._head)
                    self._head = node
                else:
                    b_element = cursor.getNext()
                    while cursor.getNext() is not None:
                        if b_element.getElement() == element:
                            break
                        cursor = cursor.getNext()
                    if cursor.getNext() is None:
                        print('new node element not found in this SLL, adding to Tail')
                        cursor.setNext(node)
                    else:
                        print(f'Inserting Before {element}')
                        node.setNext(cursor.getNext())        # set NN’s next pointer to point to x
                        cursor.setNext(node)                  # set bx’s next pointer point to NN
        self._size+=1
        
        
        
class Empty(Exception):
    """Empty exception class provided to flag that condition. """
    pass
```

#### Batch Processing Code:

```python
SLL1 = SinglyLinkedList()
with open('A4_transaction_input_stream_presidents.txt') as f:
    contents = f.read()

i = 1
contents = contents.strip()
for c in contents.split('\n'):
    print(f'process no. {i} ', end='')
    d = c.split(',')
    if d[0] == 'A':
        SLL1.add(d[1])
    elif d[0] == 'IA':
        SLL1.insertAfter(int(d[1]), d[2])
    elif d[0] == 'IE':
        SLL1.insertElt(d[1], d[2], d[3])
    elif d[0] == 'SL':
        SLL1.showList()
    else:
        SLL1.remove()
    i+=1
print('\n--------------------------\nAfter processing:\n--------------------------\n')        
SLL1.showList()
```
