### Insertion Sort:

- Code from Week 2 class content 'A1_insertion_sort.ipynb'

```python
# prepping the time library
import time
import matplotlib.pyplot as plt
```

```python
# # Load inputs
f = open("IT309-S21-A1input100.txt", "r")
A1_100 = f.read().split('\n')
del A1_100[-1]

f = open("IT309-S21-A1input1000.txt", "r")
A1_1k = f.read().split('\n')
del A1_1k[-1]

f = open("IT309-S21-A1input10000.txt", "r")
A1_10k = f.read().split('\n')
del A1_10k[-1]

f = open("IT309-S21-A1input10000s.txt", "r")
A1_10k_s = f.read().split('\n')
del A1_10k_s[-1]

f = open("IT309-S21-A1input50000.txt", "r")
A1_50k = f.read().split('\n')
del A1_50k[-1]

f = open("IT309-S21-A1input100000.txt", "r")
A1_100k = f.read().split('\n')
del A1_100k[-1]

f = open("IT309-S21-A1input200000.txt", "r")
A1_200k = f.read().split('\n')
del A1_200k[-1]

f = open("IT309-S21-A1input500000.txt", "r")
A1_500k = f.read().split('\n')
del A1_500k[-1]
```

```python
start = time.time()

def insertion_sort(A):
    swaps = 0
    k=1
    while k < len(A):
        j = k 
        while j>0 and A[j] < A[j-1]:
            A[j],A[j-1] = A[j-1],A[j]
            swaps += 1
            j-=1
        k+=1
    return A, swaps

i_sort, s = insertion_sort(A1_100)
end = time.time()
# print("sorted list:") 
# print(i_sort)
print('Elapsed Time: ', end - start, ' seconds')
print('#swaps: ', s)
```

### Merge Sort:

- Code from Week 2 class content 'M3_merge_sort.ipynb'
- Input: Integer array of varying sizes from size 100 to 500,000
- Output: A sorted list of the same size as the input

```python
# # Load inputs
f = open("IT309-S21-A1input100.txt", "r")
A1_100 = f.read().split('\n')
del A1_100[-1]

f = open("IT309-S21-A1input1000.txt", "r")
A1_1k = f.read().split('\n')
del A1_1k[-1]

f = open("IT309-S21-A1input10000.txt", "r")
A1_10k = f.read().split('\n')
del A1_10k[-1]

f = open("IT309-S21-A1input10000s.txt", "r")
A1_10k_s = f.read().split('\n')
del A1_10k_s[-1]

f = open("IT309-S21-A1input50000.txt", "r")
A1_50k = f.read().split('\n')
del A1_50k[-1]

f = open("IT309-S21-A1input100000.txt", "r")
A1_100k = f.read().split('\n')
del A1_100k[-1]

f = open("IT309-S21-A1input200000.txt", "r")
A1_200k = f.read().split('\n')
del A1_200k[-1]

f = open("IT309-S21-A1input500000.txt", "r")
A1_500k = f.read().split('\n')
del A1_500k[-1]
```

```python
start = time.time()

def merge(S1,S2,S):
    i=j=0
    while i+j < len(S):
        if j==len(S2) or (i<len(S1) and S1[i] < S2[j]):
            S[i+j] = S1[i]
            i+=1
        else:
            S[i+j]=S2[j]
            j+=1
            

def merge_sort(S):
    global ms_count, m_count
    n=len(S)
    if n<2:
        return    # list is already sorted
    # divide
    mid = n//2
    S1 = S[0:mid]   # copy of first half 
    S2 = S[mid:n]   # copy of second half 
    
    ms_count += 1
    merge_sort(S1)   # sort copy of first half
    
    merge_sort(S2)   # sort copy of second half
    ms_count += 1

    m_count += 1
    merge(S1,S2,S)  # merge sorted halves back into S
    return S

global m_count, ms_count
ms_count = 0
m_count = 0
merge_sort(A1_500k)
end = time.time()
print('Elapsed time ', end-start, ' seconds')
print ('Merge sort recursive called: ', ms_count)   
print('Merge operation preforemed: ', m_count)
```
